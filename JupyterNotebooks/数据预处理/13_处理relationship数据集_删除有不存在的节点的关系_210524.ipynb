```python
entity = pd.read_csv('entity_210507.csv',encoding = 'utf-8')
entity = entity.drop('Unnamed: 0',axis = 1)
entity
```

```python
relationship = pd.read_csv('relationship_201203.csv',encoding = 'utf-8')
relationship = relationship.drop('Unnamed: 0',axis = 1)
relationship
```

```python
a = time.time()
delete_list = []
for i in range(len(relationship)):
    if relationship[':START_ID'][i] not in entity['entityId:ID'] or relationship[':END_ID'][i] not in entity['entityId:ID']:
        delete_list.append(i)
b = time.time()
print(b-a)
```

```python
relationship = relationship.drop(index=delete_list)
```
