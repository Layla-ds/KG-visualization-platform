# Read csv files

## Relationship

```python
relationship = pd.read_csv('relationship.csv',encoding = 'utf-8')
relationship = relationship.drop('Unnamed: 0',axis = 1)
relationship
```

```python
relationship = relationship.drop_duplicates()
```

```python
relationship
```

## Entity

```python
node1 = list(set(relationship[':START_ID']))
node2 = list(set(relationship[':END_ID']))
node = node1+node2
```

```python
allnodes = list(set(node))
```

```python
df_all = pd.DataFrame(allnodes, columns = ['entityName'])
df_all
```

```python
entity = df_all[1:]
entity['entityId:ID'] = [i for i in range(len(entity))]
entity
```

```python
entity.to_csv('entity.csv')
```
