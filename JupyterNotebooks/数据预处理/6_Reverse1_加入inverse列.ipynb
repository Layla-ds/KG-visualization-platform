```python
import pandas as pd
```

```python
relationship = pd.read_csv('relationship_new.csv',encoding = 'utf-8')
relationship = relationship.drop('Unnamed: 0',axis = 1)
relationship
```

```python
inverse = relationship
inverse = inverse.rename(columns = {':START_ID':'END_ID'})
inverse = inverse.rename(columns = {':END_ID':':START_ID'})
inverse = inverse.rename(columns = {'END_ID':':END_ID'})
inverse
```

```python
inverse = inverse[[':START_ID',':TYPE',':END_ID']]
```

```python
inverse['inverse'] = ['yes' for i in range(len(inverse))]
```

```python
relationship['inverse'] = ['no' for i in range(len(relationship))]
```

```python
df = pd.concat([relationship,inverse], axis=0)
df
```

```python
df.to_csv('relationship.csv')
```
