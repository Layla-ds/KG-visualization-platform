## Subsequence Chart

```python
subs_dict = entity.set_index('entityType')[':LABEL'].to_dict()
```

```python
print(len(subs_dict))
subs_dict
```

```python
smallLabel = entity['entityType'].value_counts()
```

```python
bigLabel = entity[':LABEL'].value_counts()
```

```python
SLdic = smallLabel.to_dict()
BLdic = bigLabel.to_dict()
```

```python
from collections import Counter
SL = list(dict(Counter(SLdic)).keys())
countSL = list(dict(Counter(SLdic)).values())
dataSL = []
for i in range(len(SLdic)):
    temp = dict()
    temp['label']=SL[i]
    temp['y']=countSL[i]
    dataSL.append(temp)
dataSL
```

```python
BL = list(dict(Counter(BLdic)).keys())
countBL = list(dict(Counter(BLdic)).values())
dataBL = []
for i in range(len(BLdic)):
    temp = dict()
    temp['label']=BL[i]
    temp['y']=countBL[i]
    dataBL.append(temp)
dataBL
```
